```python
from HashTable_OpenAddressing import *
from nltk.corpus import stopwords

table = HashTable()

stopwords = set(stopwords.words('english'))
#print(stopwords)

file = open('course_catalog.txt', 'r')             #choose whatever text file you want

lines = file.readlines()

for line in lines:
    words = line.split(' ')
    for word in words:
        word = word.rstrip()
        if word in stopwords:
            pass
        else:
            if table.search(word):
                table.insert(word, table.search(word)+1)
            else:
                table.insert(word,1)
            
wordcount = {}

for key, count in table.traverse():
    wordcount[key] = count

for word, count in wordcount.items():
    print(f"{word}: {count}")
```
